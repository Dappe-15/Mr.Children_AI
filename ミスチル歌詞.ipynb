```python
import requests
import pandas as pd
import sys
import numpy as np
from bs4 import BeautifulSoup
import re
from time import sleep
from PIL import Image
from janome.tokenizer import Tokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import markovify
%matplotlib inline
```

```python
def scraping(url):
  sleep(1)
  html = requests.get(url)
  soup = BeautifulSoup(html.content, "html.parser")
  return soup
```

```python
soup = scraping("https://www.uta-net.com/artist/684/")
contents = []
contents.append(soup.find_all(href=re.compile('/song/\d+/$')))
contents.append(soup.find_all(href=re.compile('/song/\d+/$')))
contents.append(soup.find_all(class_=re.compile('td2')))
contents.append(soup.find_all(class_=re.compile('td3')))
contents.append(soup.find_all(class_=re.compile('td4')))
infomations = []
```

```python
href_list = []
for cont in contents:
  for j in cont:
    href = j.get('href')
    clas = j.get('class')
    if clas != None and not href in href_list:
       href_list.append(href)
```

```python
lyric_list = []
for i in range(len(href_list)):
  lyric_list.append(scraping("https://www.uta-net.com"+ href_list[i]).find(id='kashi_area').text)
```

```python
LYRIC_LIST = []
for l in range(len(lyric_list)):
  lyric_list[l] = lyric_list[l].replace('\\n','')
  lyric_list[l] = lyric_list[l].replace('\n','')
  lyric_list[l] = lyric_list[l].replace('\u3000','')
  lyric_list[l] = lyric_list[l].replace('・・・', '')
  lyric_list[l] = lyric_list[l].replace('...', '')
  lyric_list[l] = lyric_list[l].replace('...', '')
  lyric_list[l] = lyric_list[l].replace('(', '')
  lyric_list[l] = lyric_list[l].replace(')', '')
  lyric_list[l] = lyric_list[l].replace('「', '')
  lyric_list[l] = lyric_list[l].replace('」', '')
  lyric_list[l] = lyric_list[l].replace('…', '')
  lyric_list[l] = lyric_list[l].replace('、', '')
  lyric_list[l] = lyric_list[l].replace(',', '')
  LYRIC_LIST.append(lyric_list[l])
  LYRIC_LIST.append("\n")
```

```python
LYRIC = "".join(LYRIC_LIST)
```

```python
t = Tokenizer()
wr = t.tokenize(LYRIC, wakati = True)
word_list = []
for i in wr:
  word_list.append(i)
```

```python
word_list_str = ' '.join(word_list)
```

```python
import markovify

text_model = markovify.NewlineText(word_list_str, state_size=3)
for i in range(100):
  sentence = text_model.make_sentence(tries = 100)
  if sentence != None:
    print(f"-------{i}回目------")
    print(sentence.replace(" ",''))
```
